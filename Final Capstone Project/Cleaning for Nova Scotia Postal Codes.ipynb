```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('Data Files\ca_postal_codes.csv')
df.head()
```

```python
df.drop(['Unnamed: 5', 'Unnamed: 6', 'Province'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df_nova = df[df['Postal Code'].str.startswith('B')].reset_index(drop = True)
```

```python
df_nova.head()
```

```python
df_nova['Borough'] = df_nova['Place Name'].str.split('(')
df_nova.head()
```

```python
df_borough = df_nova['Borough']
df_new = pd.DataFrame(df_borough.values.tolist(), columns = ['Boroughs', 'Neighborhood'])
df_new.head()
```

```python
df_nova = df_nova.join(df_new, how = 'outer')
df_nova.head()
```

```python
df_nova.drop(['Place Name', 'Borough'], axis = 1, inplace = True)
```

```python
df_nova.head()
```

```python
df_nova = df_nova[['Postal Code', 'Boroughs', 'Neighborhood', 'Latitude', 'Longitude']]
df_nova.head()
```

```python
df_nova['Neighborhood'][14:] = df_nova['Boroughs'][14:]
```

```python
df_nova['Neighborhood'][0:13] = df_nova['Neighborhood'][0:13].apply(lambda x: x.rstrip(')'))
```

```python
df_nova.head()
```

```python
df_nova.to_csv('Nova Scotia Postal Code.csv')
```
